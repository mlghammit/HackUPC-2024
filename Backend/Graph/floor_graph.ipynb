```python
import networkx as nx

coord = {0: (41.3885318, 2.1120803),
         1:(41.3882911, 2.1124113),
         2:(41.3883359, 2.1122349),
         3:(41.3883082, 2.1125377),
         4: (41.3884144, 2.1123982),
         5:(41.3883980, 2.1126044),
         6:(41.3883716, 2.1125937)}

G = nx.Graph()
G.add_nodes_from(list(range(7)))
G.add_edges_from([(0, a) for a in range(1,6)])
G.add_edge(1, 4)
G.add_edge(5, 3)
G.add_edge(6, 3)

nx.draw(G)
```
